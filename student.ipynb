## Data cleaning

```python
# handle missing values
missing_values = df.isnull().sum()
missing_values
```

```python
#replacing "." with "_" in column names
df.columns = df.columns.str.replace('.', '_')
#checking the first 5 rows of the dataset after replacing "." with "_"
df.head()
```

```python
#removing columns with missing data in more than 50% of the rows
threshold = len(df) * 0.5
aviation_data= df.dropna(thresh=threshold, axis=1)
#checking shape of the dataset after removing columns with missing data in more than 50% of the rows
aviation_data.shape
#checking firts 5 rows
aviation_data.head()
```

```python
#dropping colums that are not needed for analysis
aviation_data_analysis= aviation_data.drop(columns=['Airport_Name', 'Airport_Code', 'Report_Status', 'Investigation_Type', 'Event_Date', 'Location', 'Publication_Date', ])
#checking first 5 rows of dataset after dropping columns that are not needed for analysis
aviation_data_analysis.head()
```

```python
#getting summary statistics of the dataset
aviation_data_analysis.describe(include='all')
```

```python
#droping duplicates
aviation_data_analysis.drop_duplicates(inplace=True)
#check shape after dropping duplicates
aviation_data_analysis.shape
```

```python
#checking first 5 rows of the dataset after dropping duplicates
aviation_data_analysis.head()
```

# Data Analysis
### Using the clean data set called 'aviation_data_analysis' analyse the following
* The make of planes most involved in accidents 
* The salvageability of the different makes of aircrafts
* The phase of flight during which most accidents occur
* The weather conditions under which most accidents occur

##### 1.Analysing aircrafts most involved in accidents

```python
# to analyse make of aircraft involved in the most accidents
# start by grouping the data by make of aircraft and counting the number of accidents
accidents_by_make = aviation_data_analysis.groupby(['Make']).size().reset_index(name='Accident_Count')
#sort the data by accident count in descending order
accidents_by_make = accidents_by_make.sort_values(by='Accident_Count', ascending=False)
#display the top 10 makes with the most accidents
accidents_by_make.head(10)
```

```python
#plotting the top 10 makes with the most accidents
plt.figure(figsize=(12, 6))
plt.barh(accidents_by_make['Make'].head(20), accidents_by_make['Accident_Count'].head(20), color='purple')
plt.xlabel('Number of Accidents')
plt.ylabel('Make of Aircraft')
plt.title('Accident-prone Aircrafts')
plt.show()
# to analyse the most accident-prone makes of aircraft
```

```python
#In addition, show the safest aircrafts by sorting the data by accident count in ascending order
accidents_by_make = accidents_by_make.sort_values(by='Accident_Count', ascending=True)
#display the top 10 makes with the least accidents
accidents_by_make.head(15)


```

##### conclusions on type of aircraft and accident count
###### Cessna aircrafts were the most involved in accidents. Accidents in Cessnas occured almost two time more than in the second aircrafts type. This points to Cessna being a high risk aircraft with probable design flaws that make it prone to accidents. 


##### 2. Analysing damage to an aircraft is salvageable

```python
# to analyse the likelihood of salvaging an aircraft involved in accidents based on the make of the aircraft
# start by grouping the data by make of aircraft and counting the number of salvaged and not salvaged aircrafts
salvaged_by_make =aviation_data_analysis.groupby(['Make', 'Aircraft_damage']).size()
#create a DataFrame from the grouped data
salvaged_by_make_df = salvaged_by_make.reset_index(name='Count')
#pivot the DataFrame to create a table with makes as rows and salvaged status as columns
salvaged_pivot = salvaged_by_make_df.pivot(index='Make', columns='Aircraft_damage', values='Count').fillna(0)
salvaged_pivot

```

```python
#calculate the counts of destroyed, substantial and minor damage and unknown damage
salvaged_pivot.value_counts(['Destroyed', 'Substantial', 'Minor', 'Unknown'])
#getting the make of aircraft with most destroyed, substantial, minor and unknown damage
top_10_destroyed = salvaged_pivot.nlargest(10, 'Destroyed')
top_10_substantial = salvaged_pivot.nlargest(10, 'Substantial')
top_10_minor = salvaged_pivot.nlargest(10, 'Minor')
top_10_unknown = salvaged_pivot.nlargest(10, 'Unknown')

#combine the top 10 into a single DataFrame
top_10_damage = pd.concat([top_10_destroyed, top_10_substantial, top_10_minor, top_10_unknown], axis=0).drop_duplicates()
print(top_10_damage)

#plotting the top 10 makes with the most destroyed, substantial, minor and unknown damage')])
 #
```

```python
top_10_damage.shape
```

```python
#plotting the salvageable aircrafts by make of aircraft

plt.figure(figsize=(12, 6))
plt.barh(top_10_damage.index, top_10_damage['Destroyed'], color='red', label='Destroyed')
plt.barh(top_10_damage.index, top_10_damage['Substantial'], color='orange', label='Substantial')
plt.barh(top_10_damage.index, top_10_damage['Minor'], color='yellow', label='Minor')
plt.barh(top_10_damage.index, top_10_damage['Unknown'], color='grey', label='Unknown')
plt.xlabel('Number of Aircrafts')
plt.ylabel('Make of Aircraft')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.title('Damage to aircrafts against Make of Aircraft')
plt.legend(['Destroyed', 'Substantial', 'Minor', 'Unknown'])
plt
#display the plot
plt.show()
```

##### Conclusion on damage to aircraft following an accident based on their make
###### Cessna and Piper suffered substantial damage making them very expensive to salvage or not worth salvaging at all. 

##### 3.Analysing phase of flight during which most accidents occur

```python
#start by grouping data of phase of flight and accident count
phase_of_flight= aviation_data_analysis.groupby(['Broad_phase_of_flight']).size()
phase_of_flight
```

```python
#plotting the phase of flight and accident count
plt.figure(figsize=(12, 6))
plt.bar(phase_of_flight.index, phase_of_flight.values, color='blue')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Accidents')
plt.title('Phase of Flight Vs Accident Count')
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

#display the graph
plt.show()
```

##### What is the relationship between accidents and the phase of flight? 
###### From the analysis, most accidents occur during landing and take off underscoring the need to strengthen existing Standard operating procedures to reduce errors in this phase that might cause accidents. 

##### 4. Analysis of weather conditions under which most accidents occured

```python
# to analyse weather conditions under which accidents occurred.
#start by grouping the data by weather conditions and counting the number of accidents
weather_conditions = aviation_data_analysis.groupby(['Weather_Condition']).size().reset_index(name='Accident_Count')
weather_conditions
```

```python
#plot a pie chart to present this data
plt.figure(figsize=(12, 6))
plt.pie(weather_conditions['Accident_Count'], labels=weather_conditions['Weather_Condition'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Weather Conditions Vs Accident Count')
plt.tight_layout()
#display the pie chart
plt.show()
```

##### what is the relationshipe between weather conditions and accidents
###### Most accidents occured during conditions where pilots relied on visual meteorological conditions to navigate. Despite clear wetahre and good conditions this resulted in the most accidents demonstrating the unreliability of visual cues in flight navigation. In contrast the number of accidents during overcast conditions resulted were less since pilots used the cockpit instruments and communication with air traffic control to navigate. This points to the need to enforce the use of IMC as the standard for navigation. 

##### 5. to analyse number of engines and accidents counts

```python
#to analyse the relationship between the number of engines an aircraft has and the number of accidents
#start by grouping the data by number of engines and counting the number of accidents
engines = aviation_data_analysis.groupby(['Number_of_Engines']).size().reset_index(name='Accident_Count')
#plotting the number of engines and accident count
plt.figure(figsize=(12, 6))
plt.scatter(engines['Number_of_Engines'], engines['Accident_Count'], color='green')
plt.xlabel('Number of Engines')
plt.ylabel('Number of Accidents')
plt.title('Number of Engines Vs Accident Count')
plt.grid(True)
#display the scatter plot
plt.show()
```

##### Is there a relationship between the number of engines and number of accidents?
###### From the analysis, it seems there was an inverse relationship between the number of engines and number of accidents. Most accidents occured in single engine aircrafts and was significantly reduced in aircrafts with more than 3 engines.
